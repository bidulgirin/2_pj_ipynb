## 삶의 만족도와 일자리만족도 비교

삶의만족도와 일자리만족도 데이터 중에서 삶의만족도 전체 값과 일자리만족도 전체값을 2015년부터 2024년까지의 값을 비교함

corr_value : np.float64(0.8234488306100337)

p_value : np.float64(0.04400389530124042)

소결 : 피밸류가 0.04이므로 일자리만족도와 삶의만족도는 0.8정도로 높은 상관관계를 가짐
자신의 직무를 통해 경제적 안정뿐만 아니라 자아실현, 사회적 기여 등의 비물질적인 가치도 충족시키는 경향이 높아지기 때문에 양의 상관관계가 있는 것으로 파악된다.

결론 : 다양한 지표들과 삶의 만족도를 비교분석한 결과 주관적건강상태비교, 일자리만족도, 여가생활만족도 임금격차 순서로 상관관계가 있는 것으로 파악된다.
그리고 사회적고립도, 사회단체참가율, 생활체육참여도의 경우 상관관계가 높을 것으로 생각햇지만 p밸류가 높아 유의미한 상관관계 분석을 하지 못했다.
주관적건강상태의 경우 신체적, 정신적 안녕을 제공 햐기 때문에 서로 긍정적인 영향을 주고받는 것으로 판단된다. 4가지 부분(직업, 가족, 성관계, 자기 자신)에 있어서 만족도는 심장 질환 위험을 13% 감소시키는 것과도 관계가 있었다는 연구결과도 존재한다.
임금격차의 경우 상대적 박탈감, 필수생활 유지의 비용증가, 사회 이동성 저하 등의 요인 존재 들로 인해 부정적인 영향을 주는것으로 파악된다는 논문들이 존재한다.
일자리만족도의 경우 자신의 직무를 통해 경제적 안정뿐만 아니라 자아실현, 사회적 기여 등의 비물질적인 가치도 충족시키는 경향이 높아지기 때문에 양의 상관관계가 존재한다는 한국노동연구원의 연구도 존재한다.
여가생활만족도의 경우 여가 활동을 통해 느끼는 즐거움과 성취감이 일상생활의 피로를 해소하고, 스트레스를 감소시키며, 심리적·정서적 안정을 제공하기 때문에 서로 긍정적인 영향을 주고받기 때문인것으로 판단된다.
삶의 만족도라는 변수와 다른 변수들 간의 상관관계를 분석함으로써 한국인들의 삶의만족도는 어떤 요소들과 상관관계를 높은지를 확인하고 더 행복한 개인, 행복한 사회를 만들고자 한다.
