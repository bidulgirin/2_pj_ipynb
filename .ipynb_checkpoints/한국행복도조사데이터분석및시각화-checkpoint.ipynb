```python
OECD_life_df = OECD_life_df.set_index("나라이름")
OECD_life_df.head()
```

```python
korea = OECD_life_df.loc[["한국"], :]
korea
```

```python
result = [i for i in korea.values]
```

```python
#그래프생성
fig = plt.figure()
fig.set_facecolor('white')


labels = korea.index

ax = sns.lineplot(
        color='r', # 색상
        linestyle='-', # 라인 스타일
        marker='o',
        ci = "sd"
)

for idx, label in enumerate(labels):
    x = korea.columns # 연도별
    y = [i for i in korea.values]
    
    sns.lineplot(x=x, y=y[idx], label=label, ax=ax) # label 범례 라벨
    plt.legend()


#그래프 정보 설정
# plt.xlim(40, 80) #x축 범위
#plt.ylim(0, 10) #y축 범위
plt.xlabel('연도') #x 라벨
plt.ylabel('삶의 만족도(10점만점)') #y 라벨
plt.title("한국의 삶의 만족도 그래프") #그래프 이름

#그래프 출력
plt.show()
```
