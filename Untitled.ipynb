```python
url = "https://product.kyobobook.co.kr/"
driver = webdriver.Chrome()
driver.set_window_size(1200, 1200)

# 홈페이지 열것이야
driver.get(url=url)
# 기다리는 시간
wait_time = .5

# 살짝 기다려요
time.sleep(wait_time)

keyword = "삶"
# 검색함

try : 
    # 검색창 찾아요
    element = driver.find_element(By.ID, 'searchKeyword')
    # 기다려요
    time.sleep(wait_time)
    # 키워드 입력해요
    element.send_keys(keyword)
    # 검색어가 입력되는 시간 고려
    time.sleep(wait_time)
    # 엔터키 입력해요
    element.send_keys(Keys.ENTER)
    
except Exception as e:
    print(e)

time.sleep(wait_time + 3)
try : 
    # 19세상품제외 필터를 하기위함
    filter = "19세상품제외"
    element = driver.find_element(By.className, "benefits_checkbox").find_element(By.XPATH, ".//*[contains(.,'" + filter + "')]")
    # 아아ㅏ아아아아아!!! 된다!!!!
    time.sleep(wait_time + 2)
    element.click()
    # 해당요소 클릭함
except Exception as e:
    print(e)

time.sleep(wait_time + 5)

#driver.execute_script("document.body.style.zoom='10%'")# 이 구세주다...아....아....

soup = BeautifulSoup(driver.page_source, 'html.parser')
soup = soup.body
```
